```python
def run_coypu_ee(message):
    url = 'https://event-extraction.skynet.coypu.org'
    json_obj = {'message': message, "key": "32T82GWPSGDJTKFN"}
    x = requests.post(url, json = json_obj)
    try: 
        prediction = x.json()
    except JSONDecodeError:
        prediction = {}
    return prediction.get("event type", None), [e["entity"] for e in prediction.get("event arguments", [{"entity": None}])] 

def run_vicuna_ee(message):
    m = [{"role": "user", "content": f"You are an event extractor which extract event information, which includes event type (storm, hurricane, tornado, flood, tsunami, explosion, wildfire, earthquake, pandemic, drought, shooting, inflation or out-of-scope when the abovementioned classes do not describe the following text), location mentions, causes of events if it is mentioned in the text (e.g., high temporature), consequences of events if it is mentioned in the text(e.g., number of death/injured). Here is the text: {message}. \n Please fill out the information below. \n Event type: \n Location mentions: \n Causes: \n Consequences:  "}]
    url = 'https://turbo.skynet.coypu.org/'
    json_obj = {'messages': message, "key": "M7ZQL9ELMSDXXE86", "temperature": 0.1}
    x = requests.post(url, json = json_obj)
    try: 
        prediction = x.json()
    except JSONDecodeError:
        prediction = {}
    return prediction
```

```python
unique_messages = df.loc[df['cluster_25_75'] == 2].title.unique()
event_types = []
entities = []
for title in unique_messages:
    prediction = run_coypu_ee(title)
    event_types.append(prediction[0])
    entities.append(prediction[1])
    print(title)
    print(f"Event type: {prediction[0]}")    
    print(f"Entities: {prediction[1]}")
    print("")
```
